# Bar Plot - Seaborn
### sns.barplot(x='', y='', data=df)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset("penguins")
```

```python
df.head()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df)
```

```python
sns.countplot(df["island"])
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex")
```

```python
max(df["bill_length_mm"])
```

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1)
```

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1,hue="sex",hue_order=["Female","Male"])
```

```python
# For Horizontal Bar Chart, You should have numerical perameters only
```

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1,orient="h")
```

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=df,orient="h")
```

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1,palette="Accent_r")
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1,palette="Accent_r",
           saturation=0.5,errcolor="red",errwidth=10)
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1,palette="Accent_r",
           saturation=0.5,errcolor="blue",capsize=0.5)
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1,palette="Accent_r",
           saturation=0.5,errcolor="red",hue="sex")
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1,palette="Accent_r",
           saturation=1,errcolor="blue",hue="sex",dodge=False)
```

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=df,order=order_1,palette="Accent_r",
           saturation=1,errcolor="blue",hue="sex",alpha=0.5)
```
