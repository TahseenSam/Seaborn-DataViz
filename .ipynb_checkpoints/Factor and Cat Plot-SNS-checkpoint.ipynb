# Factor Plot - Seaborn
### sns.factorplot(x = " ", y = " ", data = df , kind = " ")

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
display(df)
```

Size= Number of customers together

```python
sns.factorplot(x="total_bill",y="tip",data=df,kind="bar")
```

```python
sns.factorplot(x="size",y="tip",data=df)
# By default Point Plot
```

```python
sns.factorplot(x="size",y="tip",hue="sex",data=df)
```

UserWarning: The `factorplot` function has been renamed to `catplot`.

```python
sns.factorplot(x="size",y="tip",hue="sex",data=df, kind="bar")
```

```python
sns.factorplot(x="size",y="tip",hue="sex",data=df, kind="box")
```

# CatPlot - Seaborn
### sns.catplot(x = " ", y = " ", data = df , kind = " ")

```python
df
```

### Kinds
##### Scatter Plots
    > stripplot() - with kind="strip"
    > swarmplot() - with kind="swarm"
##### Distribution Plots
    > boxplot() - with kind="strip"
    > violinplot() - with kind="swarm"
    > boxenplot() - with kind="boxen"
##### Categorical Plots
    > pointplot() - with kind="point"
    > barplot() - with kind="bar"
    > countplot()- with kind="count"

```python
sns.catplot(x="size",y="tip",data=df,hue="sex")
```

```python
sns.catplot(x="size",y="total_bill",kind="bar",data=df)
```

```python
sns.catplot(x="size",y="total_bill",kind="bar",data=df,hue="sex")
```

```python
sns.catplot(x="size",kind="count",data=df,hue="sex")
```

```python
sns.catplot(x="size",y="tip",kind="box",data=df,hue="sex")
```

box= box+whisker
boxen=Box only

```python
sns.catplot(x="size",y="tip",kind="boxen",data=df,hue="sex")
```

```python
sns.catplot(x="size",y="tip",data=df,hue="sex",height=10)
```

```python
sns.catplot(x="day",y="tip",data=df,hue="sex",kind='bar')
```
