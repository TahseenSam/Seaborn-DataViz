# Box Plot
### sns.boxplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.describe().transpose()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=df)
```

From Above Figure: Friday is the lowest income day as compare to others

```python
sns.boxplot(y=df["total_bill"])
```

```python
sns.boxplot(x=df["total_bill"])
```

#### Hue

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="time")
```

#### Color and Color Palette

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",palette="rocket_r")
```

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",color="g")
```

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",palette="rocket_r",order=["Fri","Thur","Sun","Sat"])
```

### Mean

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",palette="rocket_r",showmeans=True)
```

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",palette="rocket_r",
            showmeans=True, meanprops={"marker":"+","markeredgecolor":"white"},
           linewidth=2)
```

Orient
Vertical=Default
Horizont=Need both the variables Of numeric data

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",palette="rocket_r",orient="v")
```

Swapping X and Y

```python
sns.boxplot(y="day",x="total_bill",data=df,hue="sex",palette="rocket_r")
```

```python
sns.boxplot(data=df,orient="h")
```
