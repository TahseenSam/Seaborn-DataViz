# CountPlot Seaborn
### sns.barplot(x="sex",y="size",data=df)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset("tips")
df
```

```python
sns.barplot(x="sex",y="size",data=df)
plt.show()
```

Number of males and Females have dinned here

```python
sns.countplot(x="sex",data=df)
```

```python
sns.countplot(x="sex",hue="smoker",data=df)
```

```python
sns.countplot(y="sex",hue="smoker",data=df)
```

```python
sns.countplot(x="sex",hue="smoker",data=df,palette="gist_heat",saturation=0.4)
```

```python
sns.countplot(x="sex",hue="smoker",data=df,palette="gist_heat")
```
