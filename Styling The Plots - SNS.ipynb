# Styling The Plots
    * Figure Styles
    * Removing Axis Spines
    * Scale and Content

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

```python
sns.boxplot(y=df["total_bill"])
plt.show()
```

# Grids

```python
sns.barplot(x="day",y="total_bill",data=df)
plt.grid()
plt.show()
```

```python
sns.set_style("dark")
sns.barplot(x="day",y="total_bill",data=df)
```

```python
sns.set_style("whitegrid")
sns.barplot(x="day",y="total_bill",data=df)
```

```python
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=df)
```

## Axes

```python
sns.set_style("white")
sns.barplot(x="day",y="total_bill",data=df)
```

```python
sns.set_style("white")
sns.barplot(x="day",y="total_bill",data=df)

sns.despine()
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(8,7))

sns.barplot(x="day",y="total_bill",data=df)

sns.despine()
plt.show()
```

## Context

```python
sns.set_style("white")
sns.set_context("poster",font_scale=0.6)

sns.barplot(x="day",y="total_bill",data=df)

sns.despine()
plt.show()
```

```python
sns.set_style("white")
sns.set_context("paper",font_scale=1.2)

sns.barplot(x="day",y="total_bill",data=df)

sns.despine()
plt.show()
```

```python
sns.set_style("white")
sns.barplot(x="day",y="total_bill",data=df, palette="cool")

sns.despine()
plt.show()
```
