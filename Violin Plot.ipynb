```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

Total Bill paid on sunday

```python
sns.violinplot(x="day",y="total_bill",data=df)
```

As per graph - Mostly customers have visisted on Thursday i.e. Size of violin
And their avg total bill is between 10 and 25

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="time")
```

```python
sns.violinplot(x="day",y="total_bill",data=df,linewidth=3,palette="Dark2",saturation=0.6)
```

```python
sns.violinplot(x="time",y="total_bill",data=df)
```

```python
sns.violinplot(x="time",y="total_bill",data=df,order=["Dinner","Lunch"])
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="sex")
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="sex",split=True)
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="sex",split=True,scale="width")
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="sex",split=True,scale="area")
```

```python
sns.violinplot(x="total_bill",y="day",data=df,hue="sex",split=True,scale="area")
```

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="sex",inner="box") #inner= {"box",quart,point,stick}
```
