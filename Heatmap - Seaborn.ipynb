```python
 x=np.linspace(2,10,8).reshape((2,4))
```

```python
x
```

```python
sns.heatmap(x,annot=True)
```

```python
y=[["a1","a2","a3","a4"],
  ["b1","b2","b3","b4"]]
sns.heatmap(x,annot=y,fmt="s") #s= string
```

```python
z={"fontsize":14,"color":"yellow"}
sns.heatmap(x,annot=y,fmt="s",annot_kws=z)
```

```python
sns.heatmap(x,annot=y,fmt="s",cbar=False)
```

```python
sns.heatmap(x,annot=y,fmt="s",xticklabels=False,yticklabels=False)
plt.show()
```

```python
ylabel=["r","e"]
xlabel=["r","o","s","e"]
sns.heatmap(x,annot=y,fmt="s",xticklabels=xlabel,yticklabels=ylabel)
```

```python
v=sns.heatmap(x,annot=y,fmt="s",xticklabels=False,yticklabels=False)
v.set(xlabel="Python",ylabel="Seaborn")
```

```python
v=sns.heatmap(x,annot=y,fmt="s",xticklabels=False,yticklabels=False)
v.set(xlabel=["Python","C++","Java","C#"],ylabel="Seaborn")
```

```python
v=sns.heatmap(x,annot=y,fmt="s",xticklabels=False,yticklabels=False)
v.set(xlabel=["Python","C++","Java","C#"],ylabel="Seaborn")
sns.set(font_scale=2.5)
plt.show
```

```python
v=sns.heatmap(x,annot=y,fmt="s",xticklabels=False,yticklabels=False)
v.set(xlabel=["Python","C++","Java","C#"],ylabel="Seaborn")
sns.set(font_scale=1)
plt.show
```
