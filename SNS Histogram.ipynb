```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset("penguins")
```

```python
df.head()
```

# Histogram

```python
plt.hist(df["flipper_length_mm"],bins=10,color='b')
plt.show()
```

# Seaborn - displot(df["data"],bins,color,kde,rug,log_scale)

```python
sns.displot(df['flipper_length_mm'])
```

```python
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],color='indigo')
```

```python
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,color='g') 
#kde= Kernel Density Estimate
```

```python
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color='indigo')
```

```python
sns.displot(df['flipper_length_mm'],kde=True,rug=True,log_scale=True)
```
