# Scatter Plot - Seaborn
### sns.scatterplot(x=" ", y=" ", data = df)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset("taxis")
```

```python
df.head()
```

```python
sns.set(style="darkgrid")
sns.scatterplot(x=df["fare"],y=df["tip"],color="red")
```

```python
sns.scatterplot(x="distance",y="fare",data=df)
```

```python
df=sns.load_dataset("penguins")
```

```python
df.head()
```

```python
sns.scatterplot(x=df["bill_depth_mm"],y=df["bill_length_mm"],hue=df["sex"])
```

```python
sns.scatterplot(x=df.head(20)["bill_depth_mm"],y=df.head(20)["bill_length_mm"],hue=df["sex"],style=df["sex"])
```

```python
df2=df.head(20)
```

```python
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=df2,hue="sex",
               size="sex")
```

```python
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=df2,hue="sex",
               style="sex",size="sex",sizes=(20,30))
```

```python
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=df2,hue="sex",
               palette="Dark2_r",size="sex")
```

```python
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=df2,hue="sex",
               palette="Dark2_r",size="sex",alpha=0.6)
```

```python
m={"Male":"+", "Female":"o"}
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=df2,hue="sex",
                markers=m)
```
