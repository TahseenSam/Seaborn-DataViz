# Strip Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset("tips")
df
```

```python
sns.countplot(df["sex"])
```

Bill on the different days

```python
sns.stripplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
df["day"].describe()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",palette="rocket_r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,linewidth=0.5,edgecolor="yellow")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,jitter=4)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,size=6,linewidth=0.5)
plt.show()
```

```python
m={"Male":"+","Female":"<"}
sns.stripplot(x="day",y="total_bill",data=df,size=6,linewidth=0.5,edgecolor="yellow",marker="<")
plt.show()
```

```python
m={"Male":"+","Female":"<"}
sns.stripplot(x="day",y="total_bill",data=df,size=6,linewidth=0.5,edgecolor="yellow",palette="Accent_r",alpha=0.7)
plt.show()
```
